```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import wine red data
wine_red=pd.read_csv('winequality-red.csv',';')
wine_red['type']='red' # create column with red
wine_red
```

```python
# import wine white data
wine_white=pd.read_csv('winequality-white.csv',';')
wine_white['type']='white' # create column with white
wine_white
```

```python
# divide train test
wine_red_train=wine_red[:1279]
wine_red_test=wine_red[1279:]
wine_white_train=wine_white[:3918]
wine_white_test=wine_white[3918:]
```

```python
# concat wine red and wine white data
train=pd.concat([wine_red_train,wine_white_train],ignore_index=True)
test=pd.concat([wine_red_test,wine_white_test],ignore_index=True)
```

```python
# encoding througt ordinalencoder
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

x_train_wine=train.drop('type',axis=1)
y_train_wine=encoder.fit_transform(train[['type']])
x_test_wine=test.drop('type',axis=1)
y_test_wine=encoder.transform(test[['type']])
```

```python
# fit ordinal model
from statsmodels.miscmodels.ordinal_model import OrderedModel
model_ordinal = OrderedModel(y_train_wine,x_train_wine).fit()
model_ordinal.summary()
```

```python
ypred_train_ordinal=model_ordinal.predict(x_train_wine) # predict train data
ypred_test_ordinal=model_ordinal.predict(x_test_wine)  # predict test data
```

```python
# calculate accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test_wine,np.round(ypred_test_ordinal.drop(0,axis=1)))
```

```python
# calculate the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test_wine,np.round(ypred_test_ordinal.drop(0,axis=1)))
```

```python
import statsmodels.api as sm
from sklearn.metrics import r2_score


# Fit Poisson regression model
model = sm.OLS(y_train_wine, x_train_wine)
model_ols = model.fit()
print(model_ols.summary())
ypred_test_ols_wine=model_ols.predict(x_test_wine)
print(r2_score(y_test_wine,ypred_test_ols_wine))
```

```python
class linear_regression:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.loss=[]
    def fit(self, X, y):

        self.weights = np.zeros(X.shape[1]) # create feature number of list of zeros
        self.bias = 0 # intialize bias

        for i in range(self.n_iterations):

            y_pred = self.predict(X) # prediction


            dw = (1 / X.shape[0]) * np.dot(X.T, (y_pred - y)) # gradient of weights
            db = (1 / X.shape[0]) * np.sum(y_pred - y)     # gradient of bias

            self.weights -= self.learning_rate * dw # update the weights
            self.bias -= self.learning_rate * db # update the bias
            self.loss.append(np.sum((y_pred-y)**2)) # calculate loss
            if i%1000==0:
                print("loss {} : {}".format(i,np.sum((y_pred-y)**2)))

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
 # convert dataframe to series
y_train_wine=y_train_wine.ravel()
y_test_wine=y_test_wine.ravel()
# fit linear regression on our own model
model_linear_regression_wine=linear_regression(learning_rate=0.0001, n_iterations=10000)
model_linear_regression_wine.fit(x_train_wine,y_train_wine)
ypred_test_linear_regression_wine=model_linear_regression_wine.predict(x_test_wine)
print(r2_score(y_test_wine,ypred_test_linear_regression_wine))
```
