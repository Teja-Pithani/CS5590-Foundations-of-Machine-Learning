```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data through pandas
food_df=pd.read_csv('food_df.csv')
food_df
```

```python
# divide the week1 and week2 data
food_df_week1 = food_df.iloc[:28]  # First week
food_df_week2 = food_df.iloc[28:]  # Second week
```

```python

# Plot time taken for each meal type against different days for both weeks

l=[food_df_week1,food_df_week2]
for i in range(1,3):
    df=l[i-1]
    meal_categories = df['Category'].unique()
    plt.figure(figsize=(12, 6))
    for category in meal_categories:
        category_df = df[df['Category'] == category]
        days = category_df['Day']
        time_to_finish = category_df['Time_to_Finish (minutes)']
        plt.plot(days, time_to_finish, label=category)

    plt.xlabel('Day')
    plt.ylabel('Time to Finish (minutes)')
    plt.title('Time Taken for Each Meal Type Against Different Days week {}'.format(i))
    plt.legend()
    plt.show()



```

```python
#b)
# Convert 'Time' to datetime and extract hours and minutes
food_df['Time'] = pd.to_datetime(food_df['Time']).dt.time
food_df['Hour'] = food_df['Time'].apply(lambda x: x.hour)
food_df['Minute'] = food_df['Time'].apply(lambda x: x.minute)
food_df

```

```python

# divide train test
from sklearn.model_selection import train_test_split

x=food_df.drop(['Time_to_Finish (minutes)','Time'],axis=1)
y=food_df['Time_to_Finish (minutes)']
x=pd.get_dummies(x,drop_first=True) # encoding throght pandas
x_train_food,x_test_food,y_train_food,y_test_food=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import statsmodels.api as sm
from sklearn.metrics import r2_score

# Fit Poisson regression model
poisson_model = sm.GLM(y_train_food, x_train_food, family=sm.families.Poisson())
poisson_results = poisson_model.fit()

# Print the summary of the regression results
print(poisson_results.summary())
```

```python
#1c)
# predict of poisson regression
ypred_test_poisson_model=poisson_results.predict(x_test_food)
# calculate the r2 score for test data
print("poisson regression test r2 score:",r2_score(y_test_food,ypred_test_poisson_model))
```

```python

# code for poisson regression
class PoissonRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations

    def factorial(self,n):
      if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
      elif n == 0:
        return 1
      else:
        return n * self.factorial(n - 1)

    def factorial_of_list(self,numbers:list):

        return [self.factorial(n) for n in numbers]

    def link_function(self, eta):
        return np.exp(eta)

    def likelihood(self, x,y, lambda_values):
        y=y.values[0::1]
        log_likelihood_value = -np.sum(lambda_values) + np.sum(y * np.log(lambda_values)) - np.sum(np.log(self.factorial_of_list(y)))
        return -log_likelihood_value

    def fit(self, X, y):
        # Add a constant term to the independent variables
        X = np.column_stack((np.ones(X.shape[0]), X))

        num_samples, num_features = X.shape

        # Initialize coefficients to zeros
        self.beta = np.zeros(num_features)

        for i in range(self.num_iterations):
            eta = np.dot(X, self.beta)
            lambda_values = self.link_function(eta)

            # Compute the gradient
            gradient = np.dot(X.T, (y - lambda_values)) / num_samples

            # Update the coefficients using gradient descent
            self.beta += self.learning_rate * gradient

            # Compute the likelihood for monitoring convergence
            likelihood = self.likelihood(X,y, lambda_values)

            if i % 10 == 0:
                print(f'Iteration {i}, Likelihood: {likelihood}')

        return self.beta

    def predict(self, X):
        # Add a constant term to the independent variables
        X = np.column_stack((np.ones(X.shape[0]), X))

        # Predict lambda_values using the learned coefficients
        eta = np.dot(X, self.beta)
        lambda_values = self.link_function(eta)

        return lambda_values


poisson_regressor = PoissonRegression(learning_rate=0.00001, num_iterations=100)
coefficients = poisson_regressor.fit(x_train_food, y_train_food)

# Print the estimated coefficients
print('Estimated Coefficients:', coefficients)
ypred_test_poisson_model_new=poisson_regressor.predict(x_test_food)

print("poisson regression test r2 score",r2_score(y_test_food,ypred_test_poisson_model_new))
```
